```python
def number_sign_in_protocol(df, conversion):
    df = df[df['conversion'] == conversion]
    number_sign = []
    for i in range(df.shape[0]):
        number_sign.append(len(df.iloc[i, 10].split(';')))
    return number_sign
```

```python
for conv in list_conv:
    print(conv, ':', number_sign_in_protocol(df_modernizated, conv), 'all:',
          sum(number_sign_in_protocol(df_modernizated, conv)), '\n')
```
